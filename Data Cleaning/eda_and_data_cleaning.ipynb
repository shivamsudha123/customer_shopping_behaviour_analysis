Fix category and Item purchased column

```python
correct_mapping = {
    #Clothing 
 "Skirt" : "Clothing", "Socks" : "Clothing", "Blouse" : "Clothing", "Shirt" : "Clothing", "Hoodie":"Clothing","Dress" : "Clothing", "Pants" : "Clothing","Jeans" : "Clothing","Shorts" : "Clothing", "Sweater" : "Clothing", "T-shirt" : "Clothing",
 #Accessories  
 "Scarf" : "Accessories" , "Belt" : "Accessories" , "Handbag" : "Accessories" , "Hat" : "Accessories" , "Sunglasses" : "Accessories" , "Jewelry" : "Accessories" , "Gloves" : "Accessories" , "T-Shirt" : "Accessories",
 #Footwear 
 "Boots" : "Footwear", "Sandals" : "Footwear","Shoes" : "Footwear", "Sneakers" : "Footwear",
 #Outwear
 "Coat" : "Outerwear", "Jacket" : "Outerwear"
}
```

Treating null values or replacing all values in Accessories Category

```python
df.loc[df['Category'] == 'Accessories', 'Size'] = "Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.loc[df['Category'] == 'Footwear', 'Size'] = "Not Available"
```

```python
df.loc[(df['Category'] == 'Clothing') & (df["Size"].isnull()),"Size"]  = df[df ["Category"] == "Clothing"]["Size"].mode() [0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.loc[(df['Category'] == 'Outerwear') & (df["Size"].isnull()),"Size"]  = df[df ["Category"] == "Outerwear"]["Size"].mode() [0]
```

```python
df.groupby('Category')['Size'].unique()
```

Treating Category Column

```python
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
```

Treating Size Column

```python
df['Size'].isnull().sum()
```

```python
df['Size'] = df.groupby('Category')['Size'].transform(
    lambda x: x.fillna(x.mode()[0]))
```

```python
df.isnull().sum()
```

Treating Review Rating Null Values

```python
df['Review Rating'].mean()
```

```python
df['Review Rating'].median()
```

```python
df['Review Rating'].mode()
```

```python
df.groupby("Item Purchased")['Review Rating'].mean()
```

```python
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df.head(16)
```

```python
df['Review Rating'] = df['Review Rating'] .round(2)
```

```python
df.head(10)
```

Treating Previous Purchases

```python
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

Treating Purchase Amount(USD) Column

```python
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

Filling Null Values of Purchase Amount (USD) with the mean of each Item Purchased 

```python
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```

Deleting Duplicate Records

```python
df = df.drop_duplicates(subset=['Customer ID'], keep ='first')
```

Renaming the Headers

```python

df.columns = ['_'.join(col.split()) for col in df.columns]


df.columns = df.columns.str.lower()
```

```python
df.columns
```

Renaming Purchase Amount(USD) to purchase_amount

```python
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
df["subscription_status"] = df["subscription_status"].map({"Yes": True,"No":False})
df["discount_applied"] = df["discount_applied"].map({"Yes": True,"No":False})
```
