Treating the null values

Handling missing values of size column using category-based imputation

```python
df['size'].unique()
```

```python
df.groupby('category')['size'].unique()
```

```python
df.loc[df['category'] == 'Accessories', 'size'] = "Free Size"
```

```python
df.groupby('category')['size'].unique()
```

```python
df.loc[df['category'] == 'Footwear', 'size'] = "Not Available"
```

```python
df.loc[(df['category'] == 'Clothing') & (df["size"].isnull()),"size"]  = df[df ["category"] == "Clothing"]["size"].mode() [0]
```

```python
df.loc[(df['category'] == 'Outerwear') & (df["size"].isnull()),"size"]  = df[df ["category"] == "Outerwear"]["size"].mode() [0]
```

```python
df.groupby('category')['size'].unique()
```

```python
df['size'].isnull().sum()
```

Treating Category Column

```python
df['category'] = df['category'].fillna(df['category'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
```

Treating Review Rating Null Values

```python
df['review_rating'].mean()
```

```python
df['review_rating'].median()
```

```python
df['review_rating'].mode()
```

```python
df.groupby("item_purchased")['review_rating'].mean()
```

```python
df['review_rating'] = df['review_rating'].fillna(df.groupby('item_purchased')['review_rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df['review_rating'] = df['review_rating'] .round(2)
```

Treating Previous Purchases

```python
df['previous_purchases'] = df['previous_purchases'].fillna(0)
```

```python
df.isnull().sum()
```

Treating Purchase Amount(USD) Column

```python
df.groupby("item_purchased")["purchase_amount_(usd)"].mean()
```

Filling Null Values of Purchase Amount (USD) with the mean of each Item Purchased 

```python
df['purchase_amount_(usd)'] = df['purchase_amount_(usd)'].fillna(df.groupby('item_purchased')['purchase_amount_(usd)'].transform('mean'))
```

```python
df.columns[df.columns.duplicated()]
```

```python
df = df.loc[:,~df.columns.duplicated()]
```

Deleting Duplicate Records

Renaming Purchase Amount(USD) to purchase_amount

```python
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

Boolean Mapping of Subscription Status and Discount Applied Columns 

```python
df["subscription_status"] = df["subscription_status"].map({"Yes": True,"No":False})
df["discount_applied"] = df["discount_applied"].map({"Yes": True,"No":False})
```
